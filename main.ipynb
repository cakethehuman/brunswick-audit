```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# We read the data first

```python
df = pd.read_csv('data.csv')
```

```python
df
```

# Top 10 attandence

```python
df["name"].value_counts().nlargest(10).plot(kind='bar',colormap='viridis')
plt.title("Attendance by Name")
plt.xlabel("Name")
plt.ylabel("Attendance")
plt.show()
```

# There top 5 kills

```python
find_avg = df[df["name"] == "Dawnftb"]
#you can put anyname not just gwen
find_avg["kills"].mean()

number_sort = find_avg["kills"].sort_values(ascending=False)

plt.bar(find_avg["date"], find_avg["kills"].sort_values(ascending=False), color='black')


        
plt.title(f"Kills by Date by {find_avg['name'].iloc[0]}")
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Kills")
plt.show()
```
